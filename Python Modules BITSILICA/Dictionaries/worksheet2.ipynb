```python
#16.Log Session Nested Dictionary from List
lst = ['a', 'b', 'c', 'd']

nested = {}
for item in reversed(lst):
    nested = {item: nested}

print(nested)
```

```python
#17.Flip Nested Dictionary
d = {'x': {'p': 1}, 'y': {'q': 2}}

swapped = {}
for outer, inner_dict in d.items():
    for inner, value in inner_dict.items():
        swapped.setdefault(inner, {})[outer] = value

print(swapped)
```

```python
#21.Remove High-Value Keys
d = {'a': 5, 'b': 10, 'c': 15,'d': 'big'}
limit = 10

new_d = {k:v for k,v in d.items() if not isinstance(v,(int,float)) or v<=limit}
print(new_d)
```

```python
#30.Largest Subset of Anagram Words
words = ['bat', 'tab', 'eat', 'tea', 'tan', 'nat']

groups = {}

for word in words:
    key = ''.join(sorted(word)) 
    if key in groups:
        groups[key].append(word)
    else:
        groups[key] = [word]

largest_size = max(len(group) for group in groups.values())
print(largest_size)
```
