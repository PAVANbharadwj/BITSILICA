```python
#(28)
import copy

data = {
    "student": "Alice",
    "subjects": ["Math", "Physics"]
}

shallow_copy = data.copy()

deep_copy = copy.deepcopy(data)

print("Original:", data)
print("Shallow:", shallow_copy)
print("Deep:", deep_copy)

data["subjects"].append("Chemistry")

print("\nAfter modifying inner list:")
print("Original:", data)
print("Shallow:", shallow_copy)  
print("Deep:", deep_copy)   
```

```python
#8. Nested dictionaries
#(29)
students = {
    'Rahul': {'age': 16, 'marks': {'math': 90, 'english': 88}},
    'Simran': {'age': 15, 'marks': {'math': 95, 'english': 92}}
}

#(30)
print(students['Rahul']['marks']['english'])

#(31)
for s,i in students.items():
    print(s,i['marks'])

#(32)
students['Simran']['marks']['science'] = 93

print(students)
```
