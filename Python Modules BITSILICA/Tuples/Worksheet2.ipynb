```python
#31. adding element sum
t1 = (1, 2, 3, 4)
t2 = (3, 5, 2, 1)
t3 = (2, 2, 3, 1)
t4 = (1, 1, 1, 1)

tuples = [t1, t2, t3, t4]

result = tuple(sum(values) for values in zip(*tuples))
print(result)
```

```python
#32.multiplying numbers

tup = (4, 3, 2, 2, -1, 18)

prod = 1
for i in tup:
    prod*=i
    
print(prod)
```

```python
#34.sorting tuple

def sort_tup(tup):
    return sorted(tup,key = lambda x : float(x[1]),reverse=True)


t = (('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5'))

print(sort_tup(t))
```

```python
#36. frequency

t = (1, 2, 2, 3,3)

freq = {}

for i in t:
    c = t.count(i)
    freq[i] = c
    
print(freq)
```

```python
#38.check the presence

t = (('Red', 'White', 'Blue'), ('Green', 'Pink', 'Purple'), ('Orange', 'Yellow', 'Lime'))
val = "White"
flag = 0
for x in t:
    for col in x:
        if col==val:
            print("Yes")
            flag = 1
            break
if(flag==0):
    print("No")
            
```

```python
#40.AND XOR

t1 = (1, 2, 3)
t2 = (2, 2, 2)
AND = []

XOR = []

for x,y in zip(t1,t2):
    AND.append(x&y)
    XOR.append(x^y)
    
print(AND)
print(XOR)
```
