```python
#20. duplicate exist or not

tup = (1,2,3,4,5,6,7,8)

seen = set()
dupl = set()

for i in tup:
    if i in seen:
        dupl.add(i)
    seen.add(i)
    
print(len(dupl)==0)
```

```python
#21.sort in ascending order
def sort_age(tup):
    return sorted(tup,key = lambda x: x[-1])

tup =  [("Alice", 25), ("Bob", 20), ("Eve", 22)]
print(sort_age(tup))
```

```python
#22. sort tuples through digits

tup =  [(1, 2), (10, 11), (3, 44)]

def total_digits(t):
    return sum(len(str(x)) for x in t)

sorted_lst = sorted(tup, key=total_digits)

print(sorted_lst)
```

```python
#23. change last val in list of tuples

tup = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]

val = 100

for i in range(len(tup)):
    t = list(tup[i])
    t[-1] = val
    tup[i]= tuple(t)
    
print(tup)
        
```

```python
#24.remove empty tuples from the list

lst =  [(), (), ('a', 'b'), ('c',)]
n = len(lst)
i=0
while i<n:
    if not lst[i]:
        lst.remove(lst[i])
        n-=1
    else:
        i+=1
        
print(lst)
```

```python
#29. Flatten tuple
t = ((1, 2), (3, 4), (5, 6))

flat = ()
for sub in t:
    for item in sub:
        flat += (item,)

print(flat)

```

```python
#30.unique element

t = ((1, 2), (2, 3), (4, 5))

unique = set()
for sub in t:
    for item in sub:
        unique.add(item)

print(unique)
```
