```python
#5.Check if two lists have at-least one element common
my_favs = ["Tom", "Jerry", "Ben 10"]
friend_favs = ["Powerpuff", "Jerry", "Scooby"]

common = set(my_favs) & set(friend_favs)

print(common)
```

```python
#6.Python program to find common elements in three lists using sets
a = ["Toy Story", "Frozen", "Moana"]
b = ["Moana", "Coco", "Frozen"]
c = ["Frozen", "Moana", "Up"]

comman = set(a) & set(b) & set(c)

print(comman)
```

```python
#10.Python program to count number of vowels using sets in given string
def count_vowels(s):
    vowels = set("aeiouAEIOU")
    count = 0
    for ch in s:
        if ch in vowels:
            count += 1
    return count

text = "Hello World"
print("Number of vowels:", count_vowels(text))


```

```python
#13.Check if a given string is binary string or not

note = "101010"
bi = set("01")
s1 = set(note)

if s1.issubset(bi):
    print("Yes")
else:
    print("No")
```

```python
#14.Python set to check if string is panagram
sentence = "The quick brown fox jumps over a lazy dog"
apl = set("abcdefghijklmnopqrstuvwxyz")

if apl.issubset(set(sentence)):
    print("pangram")
else:
    print("Not a pangram")
```

```python
#16.Heterogram
def hetero(word):
    if len(set(word))==len(word):
        return "YES"
    return "NO"

word = "hello"
print(hetero(word))
```

```python
#24. program to get all subsets of given size of a set

from itertools import combinations

kids = {"Amy", "Bob", "Cara", "Dan", "Eva"}
size = 3

subsets = list(combinations(kids, size))
for team in subsets:
    print(team)

print("\nTotal number of teams:", len(subsets))
```

```python
#25.Minimum number of subsets with distinct elements using Counter
from collections import Counter

marbles = ["red", "blue", "red", "green", "blue", "red"]

count = Counter(marbles)

frequency = Counter(marbles).most_common(1)[0][1]
print(frequency)
```
