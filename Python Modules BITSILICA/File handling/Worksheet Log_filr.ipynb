```python
#11.extract all temperature values
import re
with open("gsm_sensor_log.txt",'r') as f:
    data =f.read()
    
    pattern = r'Temp=(\d+(?:\.\d*)?)'
    
    print(re.findall(pattern,data))
```

```python
#12.min max in temp

import re
with open("gsm_sensor_log.txt",'r') as f:
    data =f.read()
    
    pattern = r'Temp=(\d+(?:\.\d*)?)'
    
    temps = re.findall(pattern,data)
    print("Min:",min(temps))
    print("Max:",max(temps))
    
```

```python
#13.Gps coordinates

import re
with open("gsm_sensor_log.txt",'r') as f,open("gps_data.txt","w") as f1:
    data =f.read()
    
    pattern = r'Lat=(?:-?\d+(?:\.\d+)?),\s*Lon=(?:-?\d+(?:\.\d+)?)'
    
    gps = re.findall(pattern,data)
    for i in gps:
        f1.write(i)
        
    print("GPS data created")
```

```python
#15.chexk for battery warnings
import re
with open("gsm_sensor_log.txt",'r') as f:
    data =f.read()
    
    pattern = r'Battery=(\d+\.\d*)'
    
    charge = re.findall(pattern,data)
    print("Battery Warning")
    for i in charge:
        if float(i) < 3.5:
            print(i)
```
