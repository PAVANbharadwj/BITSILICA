```python
#12.What is the difference between single quotes ('), double quotes ("), and triple quotes (''' or """) for strings? Are they the same data type?
'single quotes'	Define a string. Useful when the string contains double quotes inside.
"double quotes"	Define a string. Useful when the string contains single quotes inside.
'''triple quotes''' / """triple quotes"""	Define multi-line strings or docstrings. Can span multiple lines and include both single and double quotes without escaping.
```
