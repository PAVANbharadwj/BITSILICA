```python
#12. id significance
#when we try to change te values of immutable datatypes it creates another object in another memory address
#but if we try to change the value of mutable datatypes. it changes in the same memory address, and another object will not be created

a = 10
print(id(a))
b = a
a = 20
print(id(a))
print(a is b)

s = [1,2,3]
print(id(s))
c = s
s[0] = 2
print(id(s))
print(s is c)
```
