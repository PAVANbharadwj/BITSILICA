```python
#6.score grade card
def grade(n):
    if n>=90:
        return "A"
    elif (n>=80 and n<90):
        return "B"
    elif (n>=70 and n<80):
        return "C"
    else:
        return "F"
    

n = 82
Grade = grade(n)
print(Grade)
```

```python
#11.vowels count
def string_stats(s):
    vowels = 'aeiouAEIOU'
    vowels_count = 0
    consonants_count = 0
    digits_count = 0

    for char in s:
        if char.isdigit():
            digits_count += 1
        elif char.isalpha():
            if char in vowels:
                vowels_count += 1
            else:
                consonants_count += 1

    return vowels_count, consonants_count, digits_count

print(string_stats("Hello123"))
```

```python
#12.factorial
def fact(n):
    prod = 1
    for i in range(1,n+1):
        prod *= i
    return prod
    
    
print(fact(5))
```

```python
#13.recursive function for sum of list
def Sum(lst):
    if not lst:
        return 0
    return lst[0] + Sum(lst[1:])

print(Sum([1,2,3,4]))
```

```python
#14.Reverse string recursion

def reverse(str):
    if str=="":
        return ""
    return reverse(str[1:])+str[0]

print(reverse("pavan"))
        
```

```python
#15.fibo recursion
def fibo(n):
    if n==0:
        return 0
    if n==1:
        return 1
    return fibo(n-1) + fibo(n-2)

print(fibo(4))
```

```python
#16.min max
def min_max(lst):
    Min = Max = lst[0]
    for i in lst:
        if(i>Max):
            Max = i
        if(i<Min):
            Min = i
    return Min,Max

print(min_max([8,3,5,2,10]))
```
