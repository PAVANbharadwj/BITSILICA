```python
#1.Sum of elemets in list
def deep_sum(lst):
    total = 0
    for item in lst:
        if type(item)==list:
            total += deep_sum(item)
        else:
            total += item
    
    return total

print(deep_sum([1,[2,[3,4],5],6]))
```

```python
#2.duplicates removal
def duplicates_rem(lst):
    seen = set()
    duplicates = set()
    result = []
    
    for item in lst:
        if item in seen:
            if item not in duplicates:
                duplicates.add(item)
                result.append(item)
        else:
            seen.add(item)
    return result


print(duplicates_rem([1,2,1,2,3,4,5,3]))
```

```python
#3.permutations of a string
def permutations(s):
    if len(s) <= 1:
        return [s]
    
    result = []
    seen = set()
    for i,char in enumerate(s):
        if char in seen:
            continue
        seen.add(char)
        for perm in permutations(s[:i] + s[i+1:]):
            result.append(char+perm)
    return result

print(permutations("aac"))
```

```python
#4.flatten dictionary
def flatten_dict(d, parent_key='', sep='_'):
    items = {}
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, dict):
            items.update(flatten_dict(v, new_key, sep=sep))
        else:
            items[new_key] = v
    return items


print(flatten_dict({'a': 1, 'b': {'c': 2, 'd': {'e': 3}}}))
```
