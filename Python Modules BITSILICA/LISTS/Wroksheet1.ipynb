```python
#1.List Creation and Indexing

lst = ["apple", "banana","cherry"]
print(lst[1])
lst[1] = "kiwi"
print(lst)
print("len:",len(lst))
```

```python
#2.Adding and Removing Items
lst.append("orange")
lst.insert(2,"mango")
lst.remove("apple")
lst.pop()
print(lst)
lst.clear()
```

```python
#3. Looping Through Lists
nums = [1,2,3,4,5]
for i in nums:
    print(i)
n = len(nums)
i=0
while(i<n):
    print(nums[i])
    i+=1
    
nums = [x**2 for x in nums]
print(nums)
```

```python
#4.List comprehention
lst = ["apple", "banana","cherry","kiwi","mango"]
new_lst = [ch for ch in lst if "a" in ch]
print(new_lst)
lst  = [ch.upper() for ch in lst]
print(lst)
lst = [ch.replace("BANANA","ORANGE") for ch in lst]
print(lst)
```

```python
#5.  Sorting and Copying Lists
lst = [3,1,4,2,5]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
nums = lst.copy()
print(nums)
print(lst)
```

```python
#6.Joining and Extending Lists
ls1 = ['a','b','c']
ls2 = [1,2,3]
ls1.extend(ls2)

print(ls1)
print(ls2)
```

```python
#7.List Methods Practice
colors = ["red", "green", "blue", "green"]
print(colors.count("green"))
print(colors.index("blue"))
colors.reverse()
print(colors)
colors.clear()
```
