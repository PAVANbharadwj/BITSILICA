```python
#23.permutations of list
def permute(lst):
    if len(lst) == 0:
        return [[]] 
    result = []
    for i in range(len(lst)):
        current = lst[i]
        remaining = lst[:i] + lst[i+1:]
        for p in permute(remaining):
            result.append([current] + p)
    return result

lst = [1, 2, 3]
all_perms = permute(lst)
print(all_perms)
```

```python
#27.remove duplicates preserve order

lst = [1,2,2,3,4,4,5,6,5]
seen = set()

res = []

for num in lst:
    if num not in seen:
        res.append(num)
        seen.add(num)
        
        
print(res)
```

```python
#28.max sum of sequence
def max_sum_subsequence(lst):
    return sum(x for x in lst if x > 0)

lst = [2, -1, 2, 3, 4, -5]
print(max_sum_subsequence(lst))
```
