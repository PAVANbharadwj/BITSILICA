```python
#6.Error code
n = int(input())
print("Critical") if n>=1000 else print("Warning") if (n>100 and n<999) else print("info")
```

```python
#7.boolean flags
power_on,overcurrent,overvoltage = 1,1,0
if overcurrent and overvoltage:
    print("Critical failure")
elif overcurrent:
    print("Shut Down: Overcurrent")
elif overvoltage:
    print("Shut Down: Overvoltage")
elif power_on and not(overcurrent or overvoltage):
    print("system safe")
```

```python
#8.input number quadrant
n = int(input())
if (n>=0 and n<=63):
    print("Quadrant 1")
elif (n>=64 and n<=127):
    print("Quadrant 2")
elif (n>=128 and n<=191):
    print("Quadrant 3")
elif (n>=192 and n<=255):
    print("Quadrant 4")
else:
    print("Enter 0-255")
```

```python
#9.band width
n = int(input())
if (n<1000):
    print("Low band")
elif (n>=1000 and n<=9999):
    print("Mid band")
elif (n>=10000 and n<=99999):
    print("High band")
else:
    print("Out of range")
```

```python
#10.Majority high
sensor = [40,65,70]
th = 50
count = sum(r>th for r in sensor)

if(count<2):
    print("Majority Low")
else:
    print("Majority High")
```

```python
#11.even odd parity
n = int(input(),16)
bi = n
count = 0
while(bi):
    count += bi&1
    bi>>=1
if(count%2==0):
    print("Even parity")
else:
    print("Odd parity")
```

```python
#12.voltage power and current
voltage = float(input("Enter voltage (V): "))
current = float(input("Enter current (mA): "))

voltage_safe = 3.0 <= voltage <= 3.3
current_safe = 10 <= current <= 500

if voltage_safe and current_safe:
    print("Power OK")
elif not voltage_safe and not current_safe:
    print("Power Error")
elif not voltage_safe:
    print("Voltage Error")
else:
    print("Current Error")
```

```python
#13.LED status
l1,l2,l3 = 0,0,0

if l1:
    print("LED1")
if l2:
    print("LED2")
if l3:
    print("LED3")
if not(l1 or l2 or l3):
    print("All are off")
```
