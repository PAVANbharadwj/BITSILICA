```python
#15.count 1s in binary
n = int(input())
count = 0
while n:
    count += n&1
    n>>=1
print(count)
```

```python
#18.lef rotate bits
n = int(input())
k = 2
rotate = ((n<<k) | (n>>(8-k))) & 0xFF
print(rotate)
```
