```python
#13.The Shape Family: Hierarchies for Area and Perimeter

import math

class Shape:
    def area(self):
        raise NotImplementedError("Subclass must implement area()")
    
    def perimeter(self):
        raise NotImplementedError("Subclass must implement perimeter()")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.radius

class Triangle(Shape):
    def __init__(self, base, height, side1, side2, side3):
        self.base = base
        self.height = height
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
    
    def area(self):
        return 0.5 * self.base * self.height
    
    def perimeter(self):
        return self.side1 + self.side2 + self.side3

class Square(Shape):
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side ** 2
    
    def perimeter(self):
        return 4 * self.side


triangle = Triangle(6,4,6,5,5)
print("Triangle area:", triangle.area())        
print("Triangle perimeter:", triangle.perimeter()) 

    
```

```python
#14. Binary Find Tree: Smart Organization

#tree
class Node:
    def __init__(self,key):
        self.key = key
        self.left = None
        self.right = None
        
class BST:
    def __init__(self):
        self.root = None
    
    def insert(self,data):
        if self.root is None:
            self.root = Node(data)
        else:
            return self._insert_recursive(self.root,data)
    
    def _insert_recursive(self,curr,data):
        if data < curr.key:
            if curr.left is None:
                curr.left = Node(data)
            else:
                return self._insert_recursive(curr.left,data)
        else:
            if curr.right is None:
                curr.right = Node(data)
            else:
                return self._insert_recursive(curr.right,data)
            
    def search(self,data):
        return self._search_recursion(self.root,data)
    
    
    def _search_recursion(self,curr,data):
        if curr is None:
            return False
        if data == curr.key:
            return True
        if data < curr.key:
            return self._search_recursion(curr.left,data)
        else:
            return self._search_recursion(curr.right,data)
        
        
lst = [3,4,1,2,12,15,23,85]
bst = BST()
for i in lst:
    bst.insert(i)
    
print(bst.search(45))
print(bst.search(2))
print(bst.search(85))
    
    
            
```

```python
#15. stack
class Stack:
    def __init__(self):
        self.items = []  

    def push(self, item):
        self.items.append(item)
        print(f"Pushed {item}")

    def pop(self):
        if self.is_empty():
            print("Stack is empty! Cannot pop.")
            return None
        return self.items.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


stack = Stack()

stack.push(10)
stack.push(20)

print("Top element:", stack.peek())    

print("Popped element:", stack.pop()) 
print("Popped element:", stack.pop())  
print("Popped element:", stack.pop())  
```

```python
#16.LL
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LIST:
    def __init__(self):
        self.head = None
    
    def insert_end(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node
            
    def insert_begin(self,data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        
    def insert_pos(self,data,index):
        new_node = Node(data)
        temp = self.head
        for i in range(index-1):
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node
        
            
    def delete_key(self,key):
        temp = self.head
        
        if temp.data == key:
            self.head = temp.next
            return
        
        
        prev = None
        while temp and temp.data!=key:
            prev = temp
            temp = temp.next
            
        if temp is None:
            print("No data")
        else:
            prev.next = temp.next
            
    def delete_start(self):
        self.head = self.head.next
        
    def delete_end(self):
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None
        
    def delete_pos(self,index):
        temp = self.head
        for i in range(index-1):
            if temp==None:
                return 
            temp = temp.next
            
        temp.next = temp.next.next
        
    def middle(self):
        slow = self.head
        fast = self.head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        print(slow.data)
        
    def display(self):
        temp = self.head
        while temp:
            print(temp.data,end = '->')
            temp = temp.next
        print("None")
        
        
        
lst = LIST()

arr = [1,2,3,4,6,7]

for i in arr:
    lst.insert_end(i)
    


lst.insert_pos(5,4)
lst.display()
lst.delete_key(5)
lst.display()
lst.delete_end()
lst.display()
lst.delete_start()
lst.display()
lst.delete_pos(3)
lst.display()
lst.middle()
```

```python
#17. shopping cart

class Shopping:
    def __init__(self):
        self.items = {}
        
    def add_items(self,item,price,count):
        if item in self.items:
            self.items[item][1] += count
        else:
            self.items[item] = [price,count]
    
    def remove_items(self,item,count=0):
        if item not in self.items.keys():
            print("item not exist in cart")
            return
        if count >= self.items[item][1] or count==0:
            del self.items[item]
        elif count < self.items[item][1]:
            self.items[item][1] -= count
        else:
            print("item not exist in cart")
        
        
    def total(self):
        if len(self.items)==0:
            print("cart is empty")
            return
        return sum(p*c for p,c in self.items.values())
    
    def show_cart(self):
        if len(self.items)==0:
            print("cart is empty")
            return
        
        for it,v in self.items.items():
                print(f"Item : {it} ---- Quantity : {v[1]} ---- price : {v[0]*v[1]}")
        print(f"                           total amount : {self.total()}")
            
            
cart = Shopping()
cart.add_items("milk",30,3)
cart.add_items("tomato",20,4)
cart.add_items("PS5",50000,1)
cart.add_items("onion",30,4)

print(cart.show_cart())
cart.remove_items("milk",1)
cart.remove_items("tomato",2)
cart.remove_items("onion")

print(cart.total())
print(cart.show_cart())
```

```python
#18. Enhanced Stack: Show Me What’s Inside
#stack
class Stack:
    def __init__(self):
        self.items = []
        
    def push(self,data):
        self.items.append(data)
        
    def pop(self):
        if self.is_empty():
            print("the stack is empty")
            return
        print(f"element {self.items[-1]} is popped")
        self.items.pop()
        
        
    def is_empty(self):
        return (len(self.items)==0)
    
    def peek(self):
        if self.is_empty():
            print("stack is empty")
            return None
        return self.items[-1]
    
    def display(self):
        if self.is_empty():
            print("stack is empty")
            return
        print(list(reversed(self.items)))
        
        
stack = Stack()

stack.push(1)
stack.push(2)
stack.push(3)
print(stack.peek())
stack.push(4)

stack.pop()
stack.pop()
stack.pop()
stack.pop()
stack.pop()


stack.display()


 
```

```python
#19. Queue
class Queue:
    def __init__(self):
        self.items = []
        
    def enque(self,data):
        self.items.append(data)
        
    def deque(self):
        if self.is_empty():
            print("Queue is empty")
            return
        print(f"element {self.items[0]} is removed")
        self.items.pop(0)
        
    def is_empty(self):
        return (len(self.items)==0)
    
    def peek(self):
        if self.is_empty():
            print("Queue is empty")
            return
        return self.items[0]
    
    def display(self):
        if self.is_empty():
            print("Queue is empty")
            return
        print(self.items)
        
        
que = Queue()

que.enque(1)
que.enque(2)
que.enque(3)
que.enque(4)
que.enque(5)
que.enque(6)

que.display()

que.deque()
que.deque()
que.deque()

que.display()

print(que.peek())

```

```python
#20. Banking system
class Bank:
    def __init__(self,name,account_no,amount):
        self.__account_no = account_no
        self.__name = name
        self.__amount = amount
        
    def deposit(self,money):
        self.__amount += money
        
    def withdraw(self,money):
        if money>self.__amount:
            print("Less funds in account")
            return
        self.__amount -= money
        
    def balance(self):
        return self.__amount
    
    
    def account_details(self):
        print(f"Name : {self.__name}\nAccount number : {self.__account_no}\nBalance : {self.__amount}")
        
        
cust = Bank("Pavan",12345,50000)
print(cust.balance())
cust.withdraw(2500)
print(cust.balance())
cust.deposit(50000)
print(cust.balance())
cust.withdraw(10000000)
cust.account_details()
    
```
