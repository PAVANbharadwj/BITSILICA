```python
#1.print 3 digit number cumes
n = 100
Sum = 0
for i in range(n,1000):
    t = i
    Sum = 0
    while(t>0):
        r = t%10
        Sum += (r**3)
        t //=10
    if(Sum==i):
        print(i)
        
```

```python
#2.Multliplication table
n = int(input())

for i in range(1,11):
    prod = 0
    for j in range(1,i+1):
        prod += n
    print(f"{n} X {i} = {prod}")
```

```python
#3.prime Numbers
def prime(n):
    count = 0
    for i in range(2,int(n/2)+1):
        if n%i==0:
            count+=1
    if count>0:
        return False
    else:
        return True
    
n = int(input())
for i in range(2,n+1):
    if prime(i):
        print(i)
```

```python
#4.pattern printing
n = int(input())
for i in range(1,n+1):
    count = 1
    for j in range(i):
        print(count,end = " ")
        count+=1
    print(' ')
```

```python
#5.reverse order
n = int(input())
rev = 0
while(n>0):
    r = n%10
    rev = rev*10 + r
    n //=10

print(rev)
```

```python
#6.palindrome number
n = int(input())
rev = 0
t = n
while(n>0):
    r = n%10
    rev = rev*10 + r
    n //=10
print(rev)
if t==rev:
    print("Yes")
else:
    print("No")
```

```python
#7.GCD of two
a = int(input())
b = int(input())

while(a!=b):
    if(a>b):
        a = a-b
    else:
        b = b-a
        

print("GCD",a)
```

```python
#8.patern printing
n = int(input())
print("*"*n)
for i in range(n-2):
    print("*",end='')
    print(" "*(n-2),end= '')
    print("*")
print("*"*n)
```

```python
#9.sum of digits
num = int(input())

while(num>9):
    Sum = 0
    while(num>0):
        r = num%10
        Sum += r
        num //= 10
    num = Sum
    
print(num)
```

```python
#14.perfect number
n = int(input())
Sum = 0
for i in range(1,(n//2)+1):
    if(n%i==0):
        Sum+=i
if(Sum==n):
    print("Perfect Number")
else:
    print("Not perfect Number")
```

```python
#20.fibo number
n = int(input())
fib = 0
if(n==0):
    fib = 0
if(n==1):
    fib = 1
else:
    a = 0
    b = 1
    for i in range(2,n+1):
        c = a+b
        fib = c
        a = b
        b = c
        
print(fib)
```
