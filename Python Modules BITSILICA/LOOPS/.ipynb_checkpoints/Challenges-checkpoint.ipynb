```python
#1.prouct of firts two

for i in range(1000,10000):
    t = i
    sum1 = 1
    sum2 = 1
    for j in range(1,5):
        if(j<3):
            r = t%10
            sum1 *= r
            t //= 10
        else:
            r = t%10
            sum2*=r
            t //= 10
    if(sum1==sum2):
        print(i)
```

```python
#2.power number sum
for i in range(100,1000):
    power = 3
    t = i
    Sum = 0
    while(t>0):
        r = t%10
        Sum += r**power
        power-=1
        t //= 10
    if(Sum==i):
        print(i)

    
```

```python
#3.factorial sum digit
for i in range(1,1000):
    t = i
    Sum = 0
    while(t>0):
        r = t%10
        prod = 1
        for j in range(1,r+1):
            prod *= j
        Sum+=prod
        t //= 10
    if(Sum==i):
        print(i)
```
