```python
#1.kernal Panic and Oops
log = """
[12345.67] kernel panic - not syncing: Fatal exception
[12345.89] Oops: 0002 [#1] SMP
[12346.00] kernel panic - not syncing: Fatal exception
[12346.13] Oops: 0003 [#2] SMP
"""

print("Kernal Panic:",log.count("kernel panic"))
print("oops:",log.count("Oops"))
```

```python
#2.extract lines with warning
log = """
[100.123] INFO: Booting Linux
[100.456] WARNING: Low memory detected
[101.000] WARNING: CPU load high
[101.555] INFO: Boot complete
"""

lst = log.split('\n')

for i in lst:
    if "WARNING" in i:
        print(i)
```

```python
#3.top 3 most frequent keywords
log = """
[1.11] ERROR: device failed to start
[1.12] ERROR: timeout waiting for response
[1.13] WARNING: low battery
[1.14] PANIC: unrecoverable error
[1.15] device fail detected
[1.16] timeout on bus
"""
count = {}
lst = log.split()

for i in range(len(lst)):
    lst[i] = lst[i].lower()

    
for word in lst:
    if word in count:
        count[word] += 1
    else:
        count[word] = 1
c = 0
for k,v in count.items():
    print(k,v)
    c+=1
    if c==3:
        break
```

```python
#4.print PLUG UNPLUG
log = """
[22.1] usb 1-1: USB disconnect, device number 2
[22.3] usb 1-1: new full-speed USB device number 3 using xhci_hcd
[22.5] usb 2-1: reset high-speed USB device number 4 using ehci-pci
"""

lst = log.split('\n')

s1 = "USB disconnect"
s2 = "new full-speed USB device"

for char in lst:
    if ((s1 in char) or (s2 in char)):
        print(char)


```

```python
#5.error hilights
log = """
[10.00] Starting system
[10.05] Initializing drivers
[10.10] ERROR: Device not found
[10.15] Trying next device
[10.20] ERROR: Timeout waiting for response
[10.25] Reboot recommended
"""

lst = log.split('\n')

s = "ERROR"

for i in range(len(lst)):
    if s in lst[i]:
        print(lst[i-1])
        print(lst[i])
        print(lst[i+1])
        print('\n')
```
