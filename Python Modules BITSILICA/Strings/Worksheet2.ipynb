```python
#31.snake case to pascal case
def rem_char(s,i):
    return s[:i]+s[i+1:]

def pascal_case(s):
    s = s[0].upper()+s[1:]
    i = 0
    n = len(s)
    while i<n:
        if s[i]=='_':
            s = rem_char(s,i)
            s = s[:i] + s[i].upper() + s[i+1:]
            n-=1
        i+=1
    
    return s
print(pascal_case("my_var_name"))
```

```python
#42.pangram check

def pangram(s):
    alp = 'abcdefghijklmnopqrstuvwxyz'
    for char in alp:
        if char not in s:
            return False
    return True

s = "The quick brown fox jumps over the lazy dog"
print(pangram(s))
```

```python
#44.split into n
def split_len(s,n):
    lst = []
    l = len(s)
    for i in range(0,l,n):
        lst.append(s[i:i+n])
    return lst

print(split_len("abcdefgh",3))
```

```python
#extract indices of string

def find_all_positions(s, sub):
    positions = []
    i = 0
    while i <= len(s) - len(sub):
        if s[i:i+len(sub)] == sub:
            positions.append(i)
        i += 1
    return positions

text = "abracadabra"
substring = "abra"
print(find_all_positions(text, substring))  # [0, 7]
```

```python
#47.redundant string
def remove_redundant_substring(s):
    n = len(s)
    for i in range(1, n//2 + 1):
        if n % i == 0:
            sub = s[:i]
            if sub * (n//i) == s:
                return sub
    return s

def clean_list(lst):
    return [remove_redundant_substring(s) for s in lst]

strings = ["hellohello", "world", "testtesttest"]
print(clean_list(strings))
```
