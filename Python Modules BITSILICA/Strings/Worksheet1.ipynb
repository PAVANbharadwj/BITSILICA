```python
#7.least frequent character
s = "statistics"
freq = {}

for char in s:
    freq[char] = freq.get(char, 0) + 1

min_freq = min(freq.values())

least_frequent = [char for char, count in freq.items() if count == min_freq]

print(f"Least frequent character: {least_frequent[0]}")


```

```python
#10.Random generation of strs
import random
import string

def random_until_target(target):
    attempts = 0
    chars = string.ascii_lowercase  
    target_length = len(target)

    while True:
        attempts += 1
        rand_str = ''.join(random.choice(chars) for _ in range(target_length))
        if rand_str == target:
            print(f"Randomly generated '{target}' after {attempts} attempts")
            break

random_until_target("abc")
```

```python
#15.permutations of a string
def permutations(s):
    if len(s)<=1:
        return [s]
    
    seen = set()
    result = []
    for i,char in enumerate(s):
        if char in seen:
            continue
        seen.add(char)
        for perm in permutations(s[:i]+s[i+1:]):
            result.append(char+perm)
            
    return result

print(permutations("abc"))
```

```python
#16.detect urls
def extract_urls(text):
    urls = []
    i = 0
    while i < len(text):
        if text.startswith("http://", i) or text.startswith("https://", i):
            start = i
            while i < len(text) and text[i] != ' ':
                i += 1
            urls.append(text[start:i])
        else:
            i += 1
    return urls

text = "Check this link: https://openai.com and http://github.com"
print("URLs found:", extract_urls(text))
```

```python
#24.all possible ip addresses
def restore_ip_addresses(s):
    res = []
    n = len(s)

    for i in range(1, min(4, n - 2)):        # first cut
        for j in range(i + 1, min(i + 4, n - 1)):  # second cut
            for k in range(j + 1, min(j + 4, n)):  # third cut
                part1, part2, part3, part4 = s[:i], s[i:j], s[j:k], s[k:]
                
                if all(is_valid(p) for p in (part1, part2, part3, part4)):
                    res.append(f"{part1}.{part2}.{part3}.{part4}")
    return res

def is_valid(segment):
    return 0 <= int(segment) <= 255 and (segment == "0" or not segment.startswith("0"))

digits = "25525511135"
print(restore_ip_addresses(digits))
```

```python
#25.convert numeric words to numbers
def words_to_numbers(text):
    num_map = {
        "zero": "0", "one": "1", "two": "2", "three": "3", "four": "4",
        "five": "5", "six": "6", "seven": "7", "eight": "8", "nine": "9",
        "ten": "10"
    }

    words = text.split()
    for i, w in enumerate(words):
        lw = w.lower().strip(".,!?") 
        if lw in num_map:
            prefix = w[:-len(lw)] if len(lw) != len(w) else ""
            suffix = w[len(lw):] if len(lw) != len(w) else ""
            words[i] = prefix + num_map[lw] + suffix

    return " ".join(words)

sentence = "I have one apple and two oranges."
print(words_to_numbers(sentence))
```

```python
#27.frequency of consec chars
def consecutive_char_frequency(s):
    if not s:
        return {}
    
    freq = {}
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            freq[s[i - 1]] = count
            count = 1
    freq[s[-1]] = count  
    return freq

text = "aabccddd"
print(consecutive_char_frequency(text))
```

```python
#28.right rotation
def reverse(s,start,end):
    return s[:start]+s[start:end+1][::-1]+s[end+1:]

def right_rot(s,k):
    n = len(s)
    s = s[::-1]
    s = reverse(s,0,k-1)
    s = reverse(s,k,n-1)
    return s

s = "hello"
k = 2
print(right_rot(s,k))
```

```python
#29.maximum rotations to get original
def max_rot(s):
    rotated =s
    count = 0
    while True:
        rotated = rotated[-1] +  rotated[:-1]
        count += 1
        if rotated==s:
            return count
        
s = "abcabc"
print(max_rot(s))
```

```python
#30.word count
def word_freq(s):
    words = s.split()
    return {word: words.count(word) for word in set(words)}

s= "apple apple banana banana banana"
print(word_freq(s))
```
